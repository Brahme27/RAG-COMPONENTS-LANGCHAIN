## Ollama 
1) is a tool that lets you run open-source language models like LLaMA (Meta), Mistral (Mistral AI), Gemma (Google), Code LLaMA (Meta), Neural Chat (Intel), and Phi (Microsoft) locally on your computer for privacy, offline access, and easy AI experimentation.


2) OLLama supports embedding models,making it possible to build retrival augmented generation(RAG) applications that combines text prompts with existing documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings

# Similarly use can use other models which are available in Ollama Hub like "llama2", "llama2-7b", "llama2-13b", "llama2-70b", "llama2-chat", "llama2-chat-7b", "llama2-chat-13b", "llama2-chat-70b", "deepseek-r1"
embedings=(
    OllamaEmbeddings(model="deepseek-r1") ## by default it uses LLama2 model
)

embedings
```

# For converting list of Texts to vectors

```python
list_of_texts=embedings.embed_documents([
    "This is a test document.",
    "This is another test document.",
    "This is yet another test document."
])

print(list_of_texts)
```

```python
result=embedings.embed_query("im feeling so happy to work in this field")
#result
```

# Using specifically embedding models that are present in OLLama
checkout here https://ollama.com/blog/embedding-models

```python
embedings=OllamaEmbeddings(model="mxbai-embed-large")

output_vector=embedings.embed_query("i hate relationships and i hate people")
```

```python
len(output_vector)
```

| Feature                   | `mxbai-embed-large`                        | General LLMs (e.g., `deepseek-r1`, `llama2`)       |
|---------------------------|--------------------------------------------|----------------------------------------------------|
| **Purpose**               | Embedding-specific model                   | General-purpose language model                    |
| **Training Objective**    | Trained for semantic similarity            | Trained for text generation (next-token prediction)|
| **Embedding Quality**     | High semantic accuracy                     | Moderate, not optimized for similarity             |
| **Use Case Fit**          | Best for search, RAG, clustering           | Best for chat, generation, reasoning               |
| **Performance**           | Fast & efficient for embeddings            | Heavier, slower for embedding-only use             |
| **Vector Semantics**      | Vectors reflect true meaning similarity    | Vectors are less focused on meaning similarity     |
| **Ollama Support**        | Supports `/embed` endpoint only            | May support `/embed`, but not always               |
